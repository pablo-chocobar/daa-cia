```python
df = df.drop(["ZIP Code","ID"], axis = 1)
```

```python
min_class = df[df["Personal Loan"] == 1]
maj_class = df[df["Personal Loan"] == 0][:len(min_class)]
```

```python
df = pd.concat([min_class,maj_class])
```

```python
df
```

```python
y = df["Personal Loan"]
x = df.drop('Personal Loan' , axis=1) 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x= scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2, random_state=0)
```

# NN

```python
def dense(inputs,weights,bias,size = (11,1)):
    y = 0
    y = np.dot(weights,inputs.T) + bias
    # a network of single neuron for the sake of simplicity
    return y 
```

```python
def sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))
```

```python
def lossfn(y_true, y_pred):
    # loss function BinaryCrossEntropy Loss
    num = (1-y_true) * np.log(1 - y_pred + 1e-7)
    den = y_true * np.log(y_pred + 1e-7)
    return -np.mean(num+den, axis=0)
```

## Optimising weights using PSO

```python
from pso import PSO
```

```python
weights = PSO(lossfn, dense, y_train.values,x_train)
```

```python
weights.position
```

```python
from sklearn.metrics import classification_report
```

```python
pso_ypred = np.round(sigmoid(dense(x_test , weights.position[:11] , weights.position[11:])))
```

```python
print(classification_report(pso_ypred , y_test.values))
```

## Optimising weights using ACO

```python
from aco import optimize_weights
```

```python
weights = optimize_weights(x_train, y_train.values, num_ants=12, num_iterations=100, evaporation_rate=0.01, alpha=0.01, beta=1, layer = dense)
```

```python
weights
```

```python
aco_ypred = np.round(sigmoid(dense(x_train , weights[0][:11] , weights[0][11:])))
```

```python
print(classification_report(aco_ypred , y_train.values))
```

## Optimising weights using Genetic Algorithm

```python
from gen import GeneticAlgorithm
```

```python
population = np.random.uniform(-1, 1, size=(12, 12))
```

```python
ga = GeneticAlgorithm(population,y_train, x_train, dense, n_parents=12, crossover_rate=1, mutation_rate=0.001, n_generations=900)
```

```python
best_weights, best_fitness = ga.evolve()
```

```python
best_weights
```

```python
gen_ypred = np.round(sigmoid(dense(x_train , best_weights[:11], best_weights[-1])))
```

```python
print(classification_report(gen_ypred , y_train.values))
```
